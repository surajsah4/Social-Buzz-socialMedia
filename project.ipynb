## CLEANING "Content" File

```python
#Importing PANDAS

import pandas as pd
```

```python
#Imporing file & making 

content_file = pd.read_csv("Content.csv")
content_file.head()
```

```python
#making a copy of file on which we will work

content_copy = content_file.copy()
content_copy.head(50)
```

```python
content_copy.info()
```

```python
#Checking NULL data

content_copy.isnull().sum()
```

```python
content_copy
```

```python
#Deleting unwanted data

content_copy = content_copy.drop(["User ID", "URL"], axis = 1)
content_copy.head()
```

```python
content_copy.tail()
```

```python
#checking for unique data (to check wheather there is any missing or duplicate value)

content_copy['Category'].unique()
```

```python
content_copy['Category'].nunique()
```

```python
#as above we found that some of the data is duplicate in inverted coma so we need to 

content_copy['Category'] = content_copy['Category'].replace('"','', regex = True )
```

```python
content_copy['Category'].unique()
```

```python
content_copy['Category'].nunique()
```

```python
#converting all the data to lowercase from category column so that we can categories 

content_copy["Category"] = content_copy["Category"].map(lambda x: x if type(x)!=str else x.lower())
```

```python
content_copy['Category'].unique()
```

```python
content_copy['Category'].nunique()
```

## CLEANING "Reactions" table

```python
#Imporing file & making a copy

reaction_db = pd.read_csv("Reactions.csv")
reaction_db.head()
```

```python
reaction_copy = reaction_db.copy()
reaction_copy.head()
```

```python
#Checking NULL data

reaction_copy.isnull().sum()
```

```python
#Deleting rows where in TYPE column is NAN

reaction_copy = reaction_copy.dropna(axis=0, subset=['Type'])
```

```python
reaction_copy.head()
```

```python
reaction_copy.isnull().sum()
```

```python
#Changing "Type" Column name 
column_name = ["Serial No.", "Content ID", "User ID", "Reaction Type", "Datetime"]

reaction_copy.columns = column_name
```

```python
reaction_copy.head()
```

```python
#Deleting User ID column

reaction_copy = reaction_copy.drop(["User ID"], axis = 1)
```

```python
reaction_copy = reaction_copy.drop(["Serial No."], axis = 1)
```

```python
reaction_copy.head()
```

## CLEANING "ReactionTypes" file

```python
#Importing 3rd data set

reactiontype = pd.read_csv("ReactionTypes.csv")
```

```python
reactiontype_copy = reactiontype.copy()
```

```python
reactiontype_copy = reactiontype_copy.drop(reactiontype_copy.columns.values[0], axis = 1)
```

```python
reaction_copy.head()
```

```python
#changing column name so that we can join
column = ["Reaction Type", "Sentimant", "Score"]

reactiontype_copy.columns = column
```

```python
reactiontype_copy.head(50)
```

```python
reactiontype_copy["Reaction Type"].isnull().sum()
```

## DATA MODELING

```python
#Doing Left Join

merge = pd.merge(content_copy, reaction_copy, on = "Content ID", how = "inner", suffixes= ["_l", "_r"])
```

```python
merge.head(50)
```

```python
merge.isnull().sum()
```

```python
merge.to_csv("test.csv")
```

```python
final_db = pd.merge(merge, reactiontype_copy, on = "Reaction Type", how = "inner", suffixes= ["_l", "_r"])
```

```python
final_db.isnull().sum()
```

```python
final_db.tail(35)
```

```python
#Applying logic to find category wise total score

category_data = final_db.groupby("Category")["Score"].sum()
category_data.head(500)
```

```python
top_5_category = category_data.sort_values(ascending=False)
```

```python
top_5_category.head()
```

```python
top_5_category.to_csv("categoryWiseData.csv")
```
